```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig=plt.figure()
ax=fig.add_axes(Axes3D(fig))
data_3d=np.loadtxt('/data/tests/jiaxing/scientific_research_with_python_demo/scientific_research_with_python_demo/data_save/afV_Nifg_test.csv',delimiter=',')
x_orig=np.arange(1,171,1)
y_orig=np.arange(10, 100, 1)
print(x_orig.shape,y_orig.shape)
x,y=np.meshgrid(x_orig,y_orig)
z=data_3d
print(x.shape,y.shape,z.shape)
ax.plot_surface(x,y,z,cmap='jet')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()
```
